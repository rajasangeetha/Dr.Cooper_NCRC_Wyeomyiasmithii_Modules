#### This program is to generate comparitive bar graph between repititive elements of different mosquitos (Drosophila melanogaster,
#### Anopheles gambiae, Aedes aegypti, Wyeomyia smithii)

##### Author : Sangeetha Vijayam
##### Date : 4/18/2022

```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.1
fig = plt.subplots(figsize =(5, 5))

# set height of bar
dMel = [0, 6.3, 11.3]
aGam = [0.7, 3, 3.1]
aAeg = [2, 14, 10.5]
wSmi = [0.09, 19.27, 3.91]


# Set position of bar on X axis
bardMel = np.arange(len(dMel))
baraGam = [x + barWidth for x in bardMel]
baraAeg = [x + barWidth for x in baraGam]
baraSmi = [x + barWidth for x in baraAeg]

# Make the plot
plt.bar(bardMel, dMel, color ='orange', width = barWidth, edgecolor ='grey', label ='D.Melanogaster')
plt.bar(baraGam, aGam, color ='yellow', width = barWidth, edgecolor ='grey', label ='A.Gambiae')
plt.bar(baraAeg, aAeg, color ='blue', width = barWidth, edgecolor ='grey', label ='A.Aegypti')
plt.bar(baraSmi, wSmi, color ='pink', width = barWidth, edgecolor ='grey', label = 'W.smithii')

# Adding Xticks
plt.xlabel('Repetitive Elements', fontweight ='bold', fontsize = 10)
plt.ylabel('%', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(dMel))], ['SINE', 'LINE', 'LTR'])
plt.legend()
plt.show()
```
