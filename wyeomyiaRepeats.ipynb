```python
#### This program is to generate pie plot for the repititive elements of Wyeomyia smithii
#### The repeat elements are collected from repeat masker output wyeomyia.polished.fasta.tbl.txt (refer the image below)
##### Author : Sangeetha Vijayam
##### Date : 4/18/2022
```

```python
#This plot is to display pie chart for repetitve elements
from matplotlib import pyplot as plt
import numpy as np


sine= 0.09
line = 19.27
penelope = 0.47
ltrElements = 3.91
dnaTransposons = 7.56
rollingCircles = 1.03
unclassified = 33.92
others = 100 - ( sine + penelope + line + ltrElements + dnaTransposons + rollingCircles + unclassified)
data = [sine, line, ltrElements, dnaTransposons, rollingCircles, unclassified, penelope, others]
  
#RepetiveElements dataset
perc = "%"
repetitveElements = ['SINE','LINE', 'LTR elements', 'DNA transposons', 'Rolling-circles', 'Unclassified', 'Penelope', 'Others']
repetitvePercent = [str(sine)+perc,str(line)+perc, str(ltrElements)+perc, str(dnaTransposons)+perc, 
                    str(rollingCircles)+perc, str(unclassified)+perc, str(penelope)+perc, str(others)+perc]
 

 
#Creating plot
fig = plt.figure(figsize =(10, 10))
plt.pie(data, labels = repetitvePercent)
plt.legend(repetitveElements)
plt.title("Wyeomyia Repetitive Elements")

# show plot
plt.show()
```
